Introduction

Modern applications often require efficient data storage and retrieval mechanisms. Traditional brute-force approaches become slow and inefficient as data grows. To address this problem, we implemented a lightweight Database Management System (DBMS) that supports basic operations like insert, delete, search, and range queries. Our proposed solution uses a B+ Tree as the underlying data structure, which provides faster performance for large datasets due to its balanced and sorted nature. This DBMS supports multiple tables, allows for persistence, and compares the performance of B+ Tree with a brute-force list-based method.

Implementation

The core of the database system is built around the B+ Tree data structure, which efficiently supports dynamic insertions, deletions, searches, and range queries. Each node in the B+ Tree contains sorted keys and pointers to child nodes, ensuring logarithmic time complexity for operations. All actual records are stored in the leaf nodes, and internal nodes guide the traversal. We implemented key operations such as:

Insert: Places a new key-value pair into the correct leaf and splits the node if necessary.

Search: Traverses the tree from root to the appropriate leaf to locate the desired key.

Delete: Removes a key from the leaf and rebalances the tree if underflow occurs.

Range Query: Quickly retrieves a sequence of records within a given key range by scanning linked leaf nodes.

This B+ Tree-based implementation is compared against a brute-force list to highlight its performance advantages.

Performance Analysis

Benchmarking compared B+ Tree and BruteForceDB on key operations like insertion, deletion, search, and range queries. B+ Tree performed better in search and range queries, while BruteForceDB was faster for insertions due to its simpler structure. Memory usage was higher for the B+ Tree. Results were presented using graphs and tables from automated tests.

Visualization

These visualizations supported debugging and demonstrated the organization of the data structure effectively.
